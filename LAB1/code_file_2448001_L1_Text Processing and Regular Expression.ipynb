```python
### (a.) Define a string containing a paragraph as the value.

paragraph = """Artificial Intelligence (AI) is not just the future, it's the present! 
From voice assistants, chatbots, and self-driving cars, to personalized recommendations, AI is everywhere. 
However, with great power comes great responsibility: ethical concerns, bias in algorithms, and privacy risks all demand attention. 
AI is also transforming healthcare, enabling early disease detection, automating diagnostics, and optimizing treatment plans. 
In education, AI-powered systems are enhancing personalized learning, adapting to student needs in real time. 
Despite the advancements, ensuring fairness, transparency, and accountability in AI systems remains a major challenge."""

```

```python
### (b.) Write a program to print the number of total words and total unique words in the paragraph.

# Importing Python's built-in string constants (like ascii_letters, digits, punctuation, etc.)
import string 

# Converting paragraph into a list of words
# Here we would remove punctuation and convert to lowercase

cleaned_paragraph = paragraph.translate(str.maketrans('', '', string.punctuation)).lower()

# Split the cleaned paragraph into individual words
words = cleaned_paragraph.split()

# Finding the total words and unique words
total_words = len(words)
unique_words = len(set(words))

print("Total number of words:", total_words)
print("Total number of unique words:", unique_words)
```

```python
### (c.) Find the frequency of all words and also display the most and least frequent word.

# Import Counter to count word frequency
from collections import Counter

# Cleaning any punctuation and converting to lowercase
cleaned_paragraph = paragraph.translate(str.maketrans('', '', string.punctuation)).lower()

# Splitting into words
words = cleaned_paragraph.split()

# Frequency of all words
word_count = Counter(words)

print("\nWord Frequencies:")
for word in word_count:
    print(word, ":", word_count[word], end = " ")
```

```python
# Most frequent word
most_frequent = max(word_count.values())
for word in word_count:
    if word_count[word] == most_frequent:
        print("Most frequent word:")
        print(word,":",word_count[word])
        break
```

```python
# Least frequent word
print("Least frequent word:")
for word in word_count:
    if word_count[word] == 1:
        print(word, end = ",")
```

```python
### (d.)Find the longest word in the paragraph.

longest_word = ""
for word in word_count:
    if len(word) > len(longest_word):
        longest_word = word
print("The longest word in the paragraph is:", longest_word)
```

#### 2.1 Write regular expressions for the following languages.

```python
### 1. The set of all alphabetic strings;

import re

# All alphabetic strings (uppercase + lowercase allowed)
pattern1 = r'^[A-Za-z]+$'  # + to ensure at least one character

# Test cases
print(bool(re.fullmatch(pattern1, "HelloWorld")))  
print(bool(re.fullmatch(pattern1, "python")))       
print(bool(re.fullmatch(pattern1, "Hello123")))     
print(bool(re.fullmatch(pattern1, "AI@2024")))      
print(bool(re.fullmatch(pattern1, "")))            
```

```python
### 2. The set of all lower case alphabetic strings ending in a b;

# Lowercase strings ending with 'b'
pattern2 = r'^[a-z]*b$'   # * here would help with zero or more lowercase letters

# Test cases
print(bool(re.fullmatch(pattern2, "ab")))         
print(bool(re.fullmatch(pattern2, "hellob")))     
print(bool(re.fullmatch(pattern2, "b")))          
print(bool(re.fullmatch(pattern2, "abc")))      
print(bool(re.fullmatch(pattern2, "abB")))     
print(bool(re.fullmatch(pattern2, "ab1b")))        
```

```python
### 3. The set of all strings from the alphabet a,b such that each a is immediately preceded by and immediately followed by a b;


pattern3 = r'^(b|bab)+$'

# Test cases
print(bool(re.fullmatch(pattern3, "bab")))          
print(bool(re.fullmatch(pattern3, "babbabbab")))     #multiple valid bab     
print(bool(re.fullmatch(pattern3, "bbbabbabbb")))    # all a’s are inside bab 
print(bool(re.fullmatch(pattern3, "aba")))            
print(bool(re.fullmatch(pattern3, "ba")))
print(bool(re.fullmatch(pattern3, "abbab"))) 
print(bool(re.fullmatch(pattern3, "babbbababa")))  #last a is not surrounded by b
```

#### 2.2 Write regular expressions for the following languages. By “word”, we mean an alphabetic string separated from other words by whitespace, any relevant punctuation, line breaks, and so forth.

```python
### 1. the set of all strings with two consecutive repeated words (e.g., “Hum-bert Humbert” and “the the” but not “the bug” or “the big bug”);

import re

# Repeated words with space in between (case insensitive)
pattern1 = r'\b([A-Za-z]+)\s+\1\b'

# Test cases
test_cases1 = [
    "Humbert Humbert",      
    "the the",             
    "The the",              
    "the bug",              
    "big bug big"          
]

for s in test_cases1:
    print(f"{s} -> {bool(re.search(pattern1, s, re.IGNORECASE))}")   # allows matching regardless of case
```

```python
### 2. all strings that start at the beginning of the line with an integer and that end at the end of the line with a word;

# Integer at start, word at end, entire line
pattern2 = r'^\d+.*\b([A-Za-z]+)\b$'

# Test cases
test_cases2 = [
    "123 the",            
    "123 something",      
    "abc the",           
    "4567 8",             
    "1 this is a test",
    "123adc?"     #Ends with punctuation, not a word
]


for s in test_cases2:
    print(f"{s} -> {bool(re.fullmatch(pattern2, s))}")
```

```python
### 3. all strings that have both the word grotto and the word raven in them (but not, e.g., words like grottos that merely contain the word grotto);

# Checking both words exist as whole words
pattern_grotto = r'\bgrotto\b'
pattern_raven = r'\braven\b'

# Test cases
test_cases3 = [
    "the grotto and the raven",       
    "raven in the grotto",             
    "grottos and ravens",              
    "the raven flew",                  
    "grotto is silent",                
    "we saw a Raven in the grotto",    # Both words (case-insensitive)
    "the RAVEN flew with Grotto"       # Both words, different casing
]

for s in test_cases3:
    found = (re.search(pattern_grotto, s, re.IGNORECASE) and
             re.search(pattern_raven, s, re.IGNORECASE))
    print(f"{s} -> {bool(found)}")
```

```python
### 4. write a pattern that places the first word of an English sentence in aregister. Deal with punctuation.


pattern = r'\b[a-zA-Z]+\b'

test_cases4 = [
    "The sun rises.",
    "no! leading space here.",
    "  leading space here.",
    "'Single quote start.'",
    '"Double quotes around"',
    "(Parentheses here)",
    "[Brackets are fine]",
    "{Curly braces too}",
    "—Dash is also common",
    "123 start with number",       # Should skip 123 and get 'start'
    "_underscore at start",        # Should get 'underscore'
    "4you should not match digit", # Should get 'you'
    "!@# noise then word"          # Should get 'noise' if it’s a valid word
]

for s in test_cases4:
    match = re.search(pattern, s)
    first_word = match.group(0) if match else "No match"
    print(f"{s} -> {first_word}")

```

```python
# Loading the file
with open("story.txt", 'r') as file:
    text = file.read()
```

```python
# Cleaning the text

# Convert to lowercase
lower_case_text = text.lower()

# Remove punctuation
no_punct_text = lower_case_text.translate(str.maketrans('', '', string.punctuation))

# Split into words
words = no_punct_text.split()
```

```python
# Basic Preprocessing

# Total characters 
total_characters = len(text)
print("Total characters:", total_characters)

# Total words (after cleaning)
total_words = len(words)
print("Total words:", total_words)

# Unique words
total_unique_words = len(set(words))
print("Total unique words:", total_unique_words)

# Total lines
lines = text.split('\n')
print("Total lines:", len(lines))

# Total sentences
sentences = sent_tokenize(text)
print("Total sentences:", len(sentences))
```

```python
# Word Frequency Analysis

word_freq = Counter(words)

print("\n Top 10 most frequent words:")
for word, count in word_freq.most_common(10):
    print(f"{word}: {count}",end =", ")
```

```python
# Longest and Shortest Words

longest_word = ""

for word in words:
    if len(word) > len(longest_word):
        longest_word = word

print("Longest word:", longest_word)

shortest_word = None

for word in words:
    if shortest_word is None or len(word) < len(shortest_word):
        shortest_word = word

print("Shortest word:", shortest_word)
```

#### 28. Define a function percent(word, text) that calculates how often a given word occurs in a text, and expresses the result as a percentage.

```python
def percent(word, text):
    count = text.count(word)  # number of times word appears
    total = len(text)         # total number of words
    return round(100 * count / total, 2)

print("Percent of 'the' in text2:", percent('a', text2), "%")
```
